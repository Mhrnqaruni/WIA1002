```python
class PriorityQueue:
    def __init__(self):
        self.queue = []

    def __str__(self):
        return ' '.join([str(i) for i in self.queue])

    def isEmpty(self):
        return len(self.queue) == 0

    def insert(self, data):
        self.queue.append(data)

    def delete(self):
        try:
            max_val = 0
            for i in range(len(self.queue)):
                if self.queue[i] > self.queue[max_val]:
                    max_val = i
            item = self.queue[max_val]
            del self.queue[max_val]
            return item
        except IndexError:
            print("Priority queue is empty.")
            exit()

if __name__ == '__main__':
    myQueue = PriorityQueue()
    elements = [4, 8, 1, 2, 9, 6, 3, 7]

    # Insert elements into the priority queue
    for elem in elements:
        myQueue.insert(elem)

    # Display all elements in the priority queue
    print("Elements in the priority queue:")
    print(myQueue)

    # Retrieve and remove the first element
    print("\nFirst element (poll):", myQueue.delete())

    # Add new element 5 into the priority queue
    myQueue.insert(5)
    print("\nAdded element 5 to the priority queue.")

    # Convert the priority queue into an array and display
    print("\nPriority queue as an array:", list(myQueue.queue))

    # Retrieve the first element
    print("\nFirst element (peek):", myQueue.queue[0])

    # Check if the priority queue contains element "1"
    print("\nContains element '1':", 1 in myQueue.queue)

    # Get the current size of the priority queue
    print("\nCurrent size of the priority queue:", len(myQueue.queue))

    # Display while removing elements until the queue is empty
    print("\nRemoving elements from the queue:")
    while not myQueue.isEmpty():
        print(myQueue.delete())

    # Sort the priority queue in reverse order
    myQueue.queue.sort(reverse=True)
    print("\nPriority queue sorted in reverse order:", myQueue)
```

```python

# Create a Priority Queue Class:
# Start by defining a class called PriorityQueue.
# Initialize an empty list (the queue) in the constructor (__init__ method).

# Insert Elements into the Priority Queue:
# Implement a method (e.g., insert) to add elements to the queue.
# Use the append method to add elements to the end of the list.
    
# Delete the First Element from the Queue:
# Create a method (e.g., delete) to remove and return the first element from the queue.
# Check if the queue is empty before attempting to delete.

# Compute Union, Difference, and Intersection:
# Use set operations to compute these:
# Union: Combine the elements from both queues (use set union).
# Difference: Remove elements from the first queue that are not in the second queue (use set difference).
# Intersection: Find common elements between the two queues (use set intersection).

# Display Results:
# Print the results of union, difference, and intersection.

# Test with Given Elements:
# Create two instances of the PriorityQueue class.
# Insert the given elements into each queue.
# Compute and display the results.

```

```python
class PriorityQueue:
    def __init__(self):
        self.queue = []

    def insert(self, data):
        self.queue.append(data)

    def delete(self):
        if not self.isEmpty():
            return self.queue.pop(0)
        else:
            print("Priority queue is empty.")
            exit()

    def isEmpty(self):
        return len(self.queue) == 0

    def __str__(self):
        return ' '.join([str(i) for i in self.queue])

if __name__ == '__main__':
    # Given elements
    elements1 = ["George", "Jim", "John", "Blake", "Kevin", "Michael"]
    elements2 = ["George", "Katie", "Kevin", "Michelle", "Ryan"]

    # Create priority queues
    pr1 = PriorityQueue()
    pr2 = PriorityQueue()

    # Insert elements into the priority queues
    for elem in elements1:
        pr1.insert(elem)
    for elem in elements2:
        pr2.insert(elem)

    # Compute union, difference, and intersection
    union_result = list(set(pr1.queue) | set(pr2.queue))
    difference_result = list(set(pr1.queue) - set(pr2.queue))
    intersection_result = list(set(pr1.queue) & set(pr2.queue))

    # Display results
    print("Union of the two priority queues:", union_result)
    print("Difference of the two priority queues:", difference_result)
    print("Intersection of the two priority queues:", intersection_result)
```

```python
class ComparableBook:
    def __init__(self, book_id, title):
        self.book_id = book_id
        self.title = title

    def __str__(self):
        return f"Book ID: {self.book_id}, Title: {self.title}"

    def __lt__(self, other):
        # Compare based on book IDs (for priority queue ordering)
        return self.book_id < other.book_id

def insert(queue, book):
    queue.append(book)

def delete(queue):
    if not isEmpty(queue):
        return queue.pop(0)
    else:
        print("Priority queue is empty.")
        exit()

def isEmpty(queue):
    return len(queue) == 0

if __name__ == '__main__':
    # Given elements
    elements = [
        ComparableBook(1065, "Effective Java: Third Edition"),
        ComparableBook(3012, "Java: A Beginner Guide Seventh Edition"),
        ComparableBook(1097, "Learn Java in One Day and Learn It Well"),
        ComparableBook(7063, "Beginning Programming with Java (Dummies)"),
        ComparableBook(6481, "Java: Programming Basic for Absolute Beginner")
    ]

    # Create priority queue (as a list)
    pr_queue = []

    # Insert elements into the priority queue
    for elem in elements:
        insert(pr_queue, elem)

    # Display the priority queue
    while not isEmpty(pr_queue):
        print("Head Element:", delete(pr_queue))
        print("Priority queue:", pr_queue)
```
