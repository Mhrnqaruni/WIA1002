```python
# Define the class named ArrayBag.
# Set a constant for the default capacity of the bag, DEFAULT_CAPACITY, to 25.
# Create the constructor (__init__ method) to initialize the bag array and the number of entries.
# Write the getCurrentSize method to return the current number of items in the bag.
# Implement the isFull method to check if the bag is at full capacity.
# Implement the isEmpty method to check if the bag is empty.
# Write the add method to add a new entry to the bag if it’s not full.
# Write the remove method to remove and return the last item from the bag if it’s not empty.
# Write the remove_entry method to remove a specific item from the bag.
# Write the clear method to remove all items from the bag.
# Implement the getFrequencyOf method to count how many times a specific item appears in the bag.
# Implement the contains method to check if a specific item is in the bag.
# Write the toArray method to return a list of all items in the bag.
# Test the class by creating an instance and using its methods to add, remove, and check items.
```

```python
class ArrayBag:
    DEFAULT_CAPACITY = 25

    def __init__(self):
        self._bag = [None] * ArrayBag.DEFAULT_CAPACITY
        self._numberOfEntries = 0

    def getCurrentSize(self):
        return self._numberOfEntries

    def isFull(self):
        return self._numberOfEntries == ArrayBag.DEFAULT_CAPACITY

    def isEmpty(self):
        return self._numberOfEntries == 0

    def add(self, newEntry):
        if not self.isFull():
            self._bag[self._numberOfEntries] = newEntry
            self._numberOfEntries += 1
            return True
        return False

    def remove(self):
        if not self.isEmpty():
            removed_entry = self._bag[self._numberOfEntries - 1]
            self._bag[self._numberOfEntries - 1] = None
            self._numberOfEntries -= 1
            return removed_entry
        return None

    def remove_entry(self, anEntry):
        for index in range(self._numberOfEntries):
            if self._bag[index] == anEntry:
                self._bag[index] = self._bag[self._numberOfEntries - 1]
                self._bag[self._numberOfEntries - 1] = None
                self._numberOfEntries -= 1
                return True
        return False

    def clear(self):
        self._bag = [None] * ArrayBag.DEFAULT_CAPACITY
        self._numberOfEntries = 0

    def getFrequencyOf(self, anEntry):
        frequency = 0
        for index in range(self._numberOfEntries):
            if self._bag[index] == anEntry:
                frequency += 1
        return frequency

    def contains(self, anEntry):
        return anEntry in self._bag[:self._numberOfEntries]

    def toArray(self):
        return self._bag[:self._numberOfEntries]

# Test the ArrayBag class
bag = ArrayBag()

# Add some items to the bag
bag.add("apple")
bag.add("banana")
bag.add("orange")
bag.add("apple")

# Test the methods
print("Current size:", bag.getCurrentSize())
print("Is the bag full?", bag.isFull())
print("Is the bag empty?", bag.isEmpty())
print("Frequency of 'apple':", bag.getFrequencyOf("apple"))
print("Does the bag contain 'banana'?", bag.contains("banana"))
print("Contents of the bag:", bag.toArray())

# Remove an item and test again
removed_item = bag.remove()
print(f"Removed item: {removed_item}")
print("Current size after removal:", bag.getCurrentSize())
print("Contents of the bag after removal:", bag.toArray())

# Remove a specific item
bag.remove_entry("banana")
print("Removed 'banana':", not bag.contains("banana"))
print("Contents of the bag after removing 'banana':", bag.toArray())

# Clear the bag and test
bag.clear()
print("Is the bag empty after clear?", bag.isEmpty())
```

```python
class ArrayBag:
    DEFAULT_CAPACITY = 25

    def __init__(self):
        self._bag = [None] * ArrayBag.DEFAULT_CAPACITY
        self._numberOfEntries = 0

    def getCurrentSize(self):
        return self._numberOfEntries

    def isFull(self):
        return self._numberOfEntries == ArrayBag.DEFAULT_CAPACITY

    def isEmpty(self):
        return self._numberOfEntries == 0

    def add(self, newEntry):
        if not self.isFull():
            self._bag[self._numberOfEntries] = newEntry
            self._numberOfEntries += 1
            return True
        return False

    def remove(self):
        if not self.isEmpty():
            removed_entry = self._bag[self._numberOfEntries - 1]
            self._bag[self._numberOfEntries - 1] = None
            self._numberOfEntries -= 1
            return removed_entry
        return None

    def remove_entry(self, anEntry):
        for index in range(self._numberOfEntries):
            if self._bag[index] == anEntry:
                self._bag[index] = self._bag[self._numberOfEntries - 1]
                self._bag[self._numberOfEntries - 1] = None
                self._numberOfEntries -= 1
                return True
        return False

    def clear(self):
        self._bag = [None] * ArrayBag.DEFAULT_CAPACITY
        self._numberOfEntries = 0

    def getFrequencyOf(self, anEntry):
        frequency = 0
        for index in range(self._numberOfEntries):
            if self._bag[index] == anEntry:
                frequency += 1
        return frequency

    def contains(self, anEntry):
        return anEntry in self._bag[:self._numberOfEntries]

    def toArray(self):
        return self._bag[:self._numberOfEntries]

    def union(self, other_bag):
        """Combines the contents of this bag and another bag into a new bag.
        
        Args:
            other_bag (ArrayBag): The other bag to union with this bag.
        
        Returns:
            ArrayBag: A new bag containing the union of this bag and other_bag.
            
        Note:
            The union of two bags may contain duplicate items. For example, if an
            object occurs five times in one bag and twice in another, the union of
            these bags contains the object seven times.
        """
        # Create a new bag for the union
        union_bag = ArrayBag()
        
        # Add all items from this bag to the union bag
        for item in self.toArray():
            union_bag.add(item)
        
        # Add all items from the other bag to the union bag
        for item in other_bag.toArray():
            union_bag.add(item)
        
        return union_bag

# Example usage:
# Create two bags
bag1 = ArrayBag()
bag2 = ArrayBag()

# Add some items to the bags
bag1.add('a')
bag1.add('b')
bag1.add('c')
bag2.add('b')
bag2.add('b')
bag2.add('d')
bag2.add('e')

# Perform the union operation
everything = bag1.union(bag2)

# Convert the result to an array and print it
print(everything.toArray())
```

```python
class ArrayBag:
    DEFAULT_CAPACITY = 25

    def __init__(self):
        self._bag = [None] * ArrayBag.DEFAULT_CAPACITY
        self._numberOfEntries = 0

    def getCurrentSize(self):
        return self._numberOfEntries

    def isFull(self):
        return self._numberOfEntries == ArrayBag.DEFAULT_CAPACITY

    def isEmpty(self):
        return self._numberOfEntries == 0

    def add(self, newEntry):
        if not self.isFull():
            self._bag[self._numberOfEntries] = newEntry
            self._numberOfEntries += 1
            return True
        return False

    def remove(self):
        if not self.isEmpty():
            removed_entry = self._bag[self._numberOfEntries - 1]
            self._bag[self._numberOfEntries - 1] = None
            self._numberOfEntries -= 1
            return removed_entry
        return None

    def remove_entry(self, anEntry):
        for index in range(self._numberOfEntries):
            if self._bag[index] == anEntry:
                self._bag[index] = self._bag[self._numberOfEntries - 1]
                self._bag[self._numberOfEntries - 1] = None
                self._numberOfEntries -= 1
                return True
        return False

    def clear(self):
        self._bag = [None] * ArrayBag.DEFAULT_CAPACITY
        self._numberOfEntries = 0

    def getFrequencyOf(self, anEntry):
        frequency = 0
        for index in range(self._numberOfEntries):
            if self._bag[index] == anEntry:
                frequency += 1
        return frequency

    def contains(self, anEntry):
        return anEntry in self._bag[:self._numberOfEntries]

    def toArray(self):
        return self._bag[:self._numberOfEntries]

    def union(self, other_bag):
        # Create a new bag for the union
        union_bag = ArrayBag()
        
        # Add all items from this bag to the union bag
        for item in self.toArray():
            union_bag.add(item)
        
        # Add all items from the other bag to the union bag
        for item in other_bag.toArray():
            union_bag.add(item)
        
        return union_bag

    def intersection(self, other_bag):
        """
        Creates an intersection of two bags and returns a new bag containing elements that appear in both bags.
        
        Args:
            other_bag (ArrayBag): The other bag to intersect with.
        
        Returns:
            ArrayBag: A new bag containing the intersection of both bags.
            
        Note:
            The intersection does not affect the contents of the original bags.
            If an object occurs multiple times in both bags, it will appear in the intersection
            as many times as it appears in the bag with the lesser count.
        """
        # Create a new instance of ArrayBag for the intersection
        intersection_bag = ArrayBag()
        
        # Create a copy of this bag's items to track checked items
        checked_items = self.toArray()
        
        # Check each item in the other bag
        for item in other_bag.toArray():
            if item in checked_items:
                # Add the item to the intersection bag
                intersection_bag.add(item)
                # Remove the item from the checked items to handle duplicates
                checked_items.remove(item)
        
        return intersection_bag

# Create two ArrayBag instances
bag1 = ArrayBag()
bag2 = ArrayBag()

# Add some items to the bags
bag1.add('a')
bag1.add('b')
bag1.add('c')
bag2.add('b')
bag2.add('b')
bag2.add('d')
bag2.add('e')

# Perform the intersection operation
common_items_bag = bag1.intersection(bag2)

# Convert the result to an array and print it
print("Intersection of bag1 and bag2 contains:", common_items_bag.toArray())
```

```python
class ArrayBag:
    DEFAULT_CAPACITY = 25

    def __init__(self):
        self._bag = [None] * ArrayBag.DEFAULT_CAPACITY
        self._numberOfEntries = 0

    def getCurrentSize(self):
        return self._numberOfEntries

    def isFull(self):
        return self._numberOfEntries == ArrayBag.DEFAULT_CAPACITY

    def isEmpty(self):
        return self._numberOfEntries == 0

    def add(self, newEntry):
        if not self.isFull():
            self._bag[self._numberOfEntries] = newEntry
            self._numberOfEntries += 1
            return True
        return False

    def remove(self):
        if not self.isEmpty():
            removed_entry = self._bag[self._numberOfEntries - 1]
            self._bag[self._numberOfEntries - 1] = None
            self._numberOfEntries -= 1
            return removed_entry
        return None

    def remove_entry(self, anEntry):
        for index in range(self._numberOfEntries):
            if self._bag[index] == anEntry:
                self._bag[index] = self._bag[self._numberOfEntries - 1]
                self._bag[self._numberOfEntries - 1] = None
                self._numberOfEntries -= 1
                return True
        return False

    def clear(self):
        self._bag = [None] * ArrayBag.DEFAULT_CAPACITY
        self._numberOfEntries = 0

    def getFrequencyOf(self, anEntry):
        frequency = 0
        for index in range(self._numberOfEntries):
            if self._bag[index] == anEntry:
                frequency += 1
        return frequency

    def contains(self, anEntry):
        return anEntry in self._bag[:self._numberOfEntries]

    def toArray(self):
        return self._bag[:self._numberOfEntries]

    def union(self, other_bag):
        # Create a new bag for the union
        union_bag = ArrayBag()
        
        # Add all items from this bag to the union bag
        for item in self.toArray():
            union_bag.add(item)
        
        # Add all items from the other bag to the union bag
        for item in other_bag.toArray():
            union_bag.add(item)
        
        return union_bag

    def intersection(self, other_bag):
        # Create a new instance of ArrayBag for the intersection
        intersection_bag = ArrayBag()
        
        # Create a copy of this bag's items to track checked items
        checked_items = self.toArray()
        
        # Check each item in the other bag
        for item in other_bag.toArray():
            if item in checked_items:
                # Add the item to the intersection bag
                intersection_bag.add(item)
                # Remove the item from the checked items to handle duplicates
                checked_items.remove(item)
        
        return intersection_bag

    def difference(self, other_bag):
        """
        Creates a difference of two bags and returns a new bag containing elements that are unique to the first bag.
        
        Args:
            other_bag (ArrayBag): The other bag to compare with.
        
        Returns:
            ArrayBag: A new bag containing the difference of the first bag minus the second bag.
            
        Note:
            The difference does not affect the contents of the original bags.
            If an object occurs multiple times in the first bag and also in the second, 
            it will appear in the difference as many times as it exceeds the count in the second bag.
        """
        # Create a new instance of ArrayBag for the difference
        difference_bag = ArrayBag()
        
        # Create a copy of the other bag's items to track checked items
        other_items = other_bag.toArray()
        
        # Check each item in this bag
        for item in self.toArray():
            if item not in other_items:
                # Add the item to the difference bag if it's not in the other bag
                difference_bag.add(item)
            else:
                # Remove the item from the other_items list to handle duplicates
                other_items.remove(item)
        
        return difference_bag
# Create two ArrayBag instances
bag1 = ArrayBag()
bag2 = ArrayBag()

# Add some items to the bags
bag1.add('a')
bag1.add('b')
bag1.add('c')
bag2.add('b')
bag2.add('b')
bag2.add('d')
bag2.add('e')

# Perform the difference operation
leftOver1 = bag1.difference(bag2)
leftOver2 = bag2.difference(bag1)

# Convert the results to arrays and print them
print("Difference of bag1 from bag2 contains:", leftOver1.toArray())
print("Difference of bag2 from bag1 contains:", leftOver2.toArray())
```
